```python
import time
def generate_news(user_question):
    sys_prompt = "你是一個專業的運動新聞編輯，我會給你文字轉播資料，請幫我用繁體中文撰寫新聞稿。"
    question = sys_prompt + user_question

    # 等待5秒再發送請求
    time.sleep(5)

    response = model.generate_content(question)

    return response.text


df['generated_text'] = df['content'].apply(generate_news)
```
